#Análise de Margem de Lucro

##Agrupamento de dados

```python
#Por segmento:
columns = ['Valor Total c/ Desconto', 'Custo Total', 'Lucro', 'Margem Lucro %']
new_df = df.groupby(by='Segmento')[columns].sum()

new_df.round(2)
```

```python
#Por País

new_df = df.groupby('País')[columns].sum()
new_df
```

```python
#Por Produto

new_df = df.groupby('Produto')[columns].sum()
new_df
```

##Adição/Remoção de variáveis


```python
df['Margem Lucro %'] = ((df['Lucro'] *100)/df['Valor Total c/ Desconto'].sum()).round(2)
df
```

##Funções de agregação (Groupby)

```python
# columns = ['Valor Total c/ Desconto', 'Custo Total', 'Lucro', 'Margem Lucro %']
df.groupby('Segmento').agg(
    {
        'Valor Total c/ Desconto': 'sum',
        'Custo Total': 'mean',
        'Lucro': 'min',
        'Margem Lucro %': ['min', 'max']
    }
).round(2)
```

##Query (filtro de dados)

```python
df_loss = df.query("Lucro < 0")
df_loss = df_loss.groupby('País')["Margem Lucro %"].sum()
```

```python
ax = df_loss.plot(kind='bar', color='lightseagreen', title='Prejuízo', ylabel='% prejuízo', xlabel='País')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
```
