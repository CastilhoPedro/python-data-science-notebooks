```python
import pandas as pd
import numpy as np
import matplotlib


np.random.seed(2)


x = np.random.rand(100)
y = 0.2 * x + 0.05 * np.random.rand(100)

# Criar o DataFrame
df = pd.DataFrame({'x': x, 'y': y})

matplotlib.pyplot.scatter(df['x'], df['y'], marker= 'o')
matplotlib.pyplot.show()
```
