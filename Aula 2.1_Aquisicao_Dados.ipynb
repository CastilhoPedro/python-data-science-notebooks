# Sobre HTML

  Caso haja um elemento web **table** em uma página HTML qualquer, o pandas é capaz de processar esse elemento e transformar em uma lista de Dataframes. Bastante útil caso queiramos retirar uma tabela de uma página web.


```python
df_html = pd.read_html('base_vendas.html')
sub_df = df_html[0]
type(sub_df)
sub_df.head(1)
```
