#Data Prep


##Padronização nomes colunas

```python
df.columns
```

```python
listnames = []
for column in df.columns:
  listnames.append(column.lower().replace(' ', '_').replace('ê', 'e').replace('ç', 'c').replace('õ', 'o').replace('á', 'a').replace('ã', 'a').replace('?', '').replace('â', 'a'))
df.columns = listnames
```

##Conversão tipos de dados

```python
df.info()
#saiu da empresa: bool
#distancia do trabalho: float
#salario: float
#hora extra: bool
#perc de aumento: float


#saiu da empresa: bool
df.saiu_da_empresa = df.saiu_da_empresa.case_when(
    [
     (df.saiu_da_empresa == 'Sim', True),
     (df.saiu_da_empresa == 'Não', False)
    ]
)
df.saiu_da_empresa = df.saiu_da_empresa.astype(bool)


#distancia do trabalho: float
df.distancia_do_trabalho = df.distancia_do_trabalho.astype(float)


#salario: float
df.salario = df.salario.astype(float)


#hora extra: bool
# df.faz_hora_extras = df.faz_hora_extras.case_when(
#     [
#      (df.faz_hora_extras == 'Sim', True),
#      (df.faz_hora_extras == 'Não', False)
#     ]
# )
# df.faz_hora_extras = df.faz_hora_extras.astype(bool)


#perc de aumento: float
df.perc_de_aumento = df.perc_de_aumento.astype(float)
```
