```python
import pandas as pd
```

```python
df = pd.read_csv('vendas_ficticias.csv')
```

```python
df.Preço = df.Preço.str.replace("'", '').astype(float)
```

```python
df.info()
```

```python
df.Data_Venda = pd.to_datetime(df.Data_Venda, format='%Y-%m-%d')
```

```python
df.columns = [i.lower() for i in df.columns]
```

```python
df.groupby(['vendedor']).count()
```

```python
df.vendedor.fillna('Vendedor não cadastrado', inplace=True)
df.produto.fillna('Produto não cadastrado', inplace=True)
df.categoria.fillna('Categoria não cadastrada', inplace=True)
```

```python
df['Margem contribuição'] = df.Faturamento/df.Faturamento.sum() * 100
```

```python
df.drop(['data_venda', 'categoria', 'vendedor'], axis=1).groupby(['produto']).sum()
```

```python
df['Ano'] = df.data_venda.dt.year
df['Mês'] = df.data_venda.dt.month_name()
```

```python
df.Ano.unique()
```

```python
df_resume = df.groupby(['Ano', 'Mês']).agg(
    {'faturamento': 'sum', 'quantidade': 'sum'}
)
```
