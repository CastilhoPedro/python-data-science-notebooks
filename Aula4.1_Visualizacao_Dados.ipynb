```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.species.unique()
```

##Gráfico de Linhas

```python
#Gráfico simples
plt.figure(figsize=(10,5))
plt.plot(df.index, df.petal_width)
plt.title('Largura sépala x Espécie')
plt.ylabel('Largura (cm)')
plt.legend(['Largura pétala'])
```

```python
#Múltiplas linhas

plt.figure(figsize=(15,5))
plt.plot(df.index, df.petal_width, color= 'gray', linewidth=.5)
plt.plot(df.index, df.petal_length, linestyle= '--', marker= 'o')
plt.legend(['Largura', 'Comprimento'])
```

##Gráfico de barras

```python
#vertical
plt.bar(df.species, df.index)
```

```python
#horizontal
plt.barh(df.species, df.index)
```

```python
sns.barplot(df, x='species', y='petal_width')
```

##Gráfico de pizza

```python
df_contagem = df.species.value_counts()
```

```python
plt.figure(figsize= (8,8))
plt.pie(df_contagem, labels=df_contagem.values, autopct='%1.1f', explode=[0, 0, 0.1])
plt.legend(df_contagem.index)
```

##Gráfico de dispersão

```python
plt.scatter(df.petal_length, df.petal_width)
```
